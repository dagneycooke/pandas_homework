```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

```python
# calculate the total number of players by looking for unique usernames
total_players = len(purchase_data["SN"].unique())

# create a data frame to store the info and then display it
player_summary = pd.DataFrame({"Total Players": [total_players]})

# display data frame
player_summary.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# calculate total number of unique items by their item ID
unique_items = len(purchase_data["Item ID"].unique())

# calculate average price of items
average_price = purchase_data["Price"].mean()

# calculate total number of purchases
total_purchases = len(purchase_data["Purchase ID"])

# calculate total revenue
total_revenue = purchase_data["Price"].sum()

# create a data frame to store the info and then display it
total_analysis = pd.DataFrame({"Number of Unique Items": [unique_items],
                               "Average Price": [average_price],
                               "Number of Purchases": [total_purchases],
                               "Total Revenue": [total_revenue]})

# format average price and total revenue to have two decimal places and a dollar sign
total_analysis["Average Price"] = total_analysis["Average Price"].astype(float).map("${:,.2f}".format)
total_analysis["Total Revenue"] = total_analysis["Total Revenue"].astype(float).map("${:,.2f}".format)

# display data frame
total_analysis.head()
```

* Percentage and Count of Other / Non-Disclosed

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# set up age bins
age_bins = [0,9,14,19,24,29,34,39,100]

# assign bin names
group_names = ["<10", "10-14", "15-19", "20-24", "25-29","30-34","35-39","40+"]

# assign bins in new column to gender_data where all duplicates users have already been removed
gender_data["Age Ranges"] = pd.cut(gender_data["Age"], age_bins, labels=group_names)

# group by bin
age_group = gender_data.groupby("Age Ranges")

# count players in each bin
age_count = age_group.count()

# copy one column of count data into list
age_bin_data = age_count["Purchase ID"].tolist()

# calculate percentage values for entire list
age_bin_percent = [i/sum(age_bin_data) for i in age_bin_data]


# combine total count and percent of players into one list
age_values =  list(zip(age_bin_data, age_bin_percent))

# create data frame to show total count and percent of players
age_summary = pd.DataFrame(age_values,
                           columns=["Total Count","Percentage of Players"],
                           index=[group_names])

# format percents correctly
age_summary["Percentage of Players"] = age_summary["Percentage of Players"].astype(float).map("{:,.2%}".format)

# need to specify how many rows to show for all 8 to be seen
age_summary.head(8)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# group by username
player_data = purchase_data.groupby("SN")

# count how many purchase made by each player
player_purchase_count = player_data.count()
player_purchase_count_data = player_purchase_count["Purchase ID"].tolist()

# calculate total amount purchased by player and sort it alphabetically
player_data_summed = purchase_data.groupby("SN").aggregate({"Price": sum})
player_total_purchase_value = player_data_summed["Price"].tolist()

# combine values 
player_average_price = []
i = 0
for i in range(0, len(player_purchase_count_data)):
    player_average_price.append(player_total_purchase_value[i]/player_purchase_count_data[i])

# create list of names for index and sort them alphabetically
spender_names = purchase_data["SN"].unique()
spender_names.sort()

# combine values for table
spender_values =  list(zip(player_purchase_count_data,player_average_price,player_total_purchase_value))

# create data table
spender_summary = pd.DataFrame(spender_values,
                           columns=["Purchase Count","Average Purchase Price","Total Purchase Value"],
                              index = [spender_names])
# sort by total purchase value
spender_summary = spender_summary.sort_values(by="Total Purchase Value",ascending=False)

# format values correctly
spender_summary["Average Purchase Price"] = spender_summary["Average Purchase Price"].astype(float).map("${:,.2f}".format)
spender_summary["Total Purchase Value"] = spender_summary["Total Purchase Value"].astype(float).map("${:,.2f}".format)

# add name to index column
spender_summary.index.set_names("SN",inplace=True)
spender_summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# grab Item ID, Item Name and Price from main data frame
popular_data = purchase_data[["Item ID","Item Name","Price"]]
popular_data = popular_data.sort_values(by="Item ID")

# group by Item ID
popular_data_group = popular_data.groupby("Item ID")

# count items in group
popular_counts = popular_data_group.count()

# save counts to list
popular_counts_data = popular_counts["Item Name"].tolist()

# drop all duplicate itmes
popular_data_subset = popular_data.copy()
popular_data_subset.drop_duplicates(subset ="Item ID", inplace = True) 
item_ID = popular_data_subset["Item ID"].tolist()
item_name = popular_data_subset["Item Name"].tolist()
popular_counts_prices = popular_data_subset["Price"].tolist()

i = 0
popular_total_value = []
for i in range(0,len(item_name)):
    popular_total_value.append(popular_counts_prices[i]*popular_counts_data[i])

# zip column values together
popular_values = list(zip(popular_counts_data,popular_counts_prices,popular_total_value))

# create data frame with two indices
popular_summary = pd.DataFrame(popular_values,
                           columns=["Purchase Count","Purchase Price","Total Purchase Value"],
                              index = [item_ID,item_name])

# sort by total purchase value
popular_summary = popular_summary.sort_values(by="Purchase Count",ascending=False)

# add number formatting
popular_summary["Purchase Price"] = popular_summary["Purchase Price"].astype(float).map("${:,.2f}".format)
popular_summary["Total Purchase Value"] = popular_summary["Total Purchase Value"].astype(float).map("${:,.2f}".format)
popular_summary.index.set_names(["Item ID","Item Name"],inplace=True)
# display the data frame
popular_summary.head()
```
